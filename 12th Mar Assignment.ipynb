# Q1

```python
"""
Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.
"""
```

```python
import scipy.stats as stats

data = [50]  # sample data
n = len(data)  # sample size
mean = sum(data) / n  # sample mean
std_dev = 5  # population standard deviation

# Calculate the standard error
std_error = std_dev / (n ** 0.5)

# The 95% confidence Interval can be given as mean Â± 1.96*SE
margin_error = 1.96 * std_error
# Calculate the confidence interval
lower_bound = mean - margin_error
upper_bound = mean + margin_error

print("The 95% confidence interval is: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

```python
"""
Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 
20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. 
Use Python to perform the test with a significance level of 0.05.
"""
```

```python
import scipy.stats as stat
import numpy as np

observed_data = [8, 7, 9, 5, 10, 8]
expected_data = [0.2*47, 0.2*47, 0.2*47, 0.1*47, 0.1*47, 0.2*47]

## Chi square Goodness OF Fit
chisquare_test_statistics,p_value=stat.chisquare(observed_data,expected_data)

significance=0.05
dof=len(observed_data)-1
critical_value=stat.chi2.ppf(1-significance,dof)

if chisquare_test_statistics > critical_value:
    print("Reject the null hypothesis")
else:
    print("Accept the null hypothesis")
```

# Q4

```python
"""
A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. 
Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.
"""
```

```python
import math
import scipy.stats as stats

n = 500
x = 60
p = x/n
se = math.sqrt(p*(1-p)/n)
cv = stats.norm.ppf(0.975)
me = cv * se
ci = (p - me, p + me)

print("95% Confidence Interval: ", ci)
```

# Q5

```python
"""
Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. 
Interpret the results.
"""
```

```python
from scipy import stats

data = [75] # sample data
n = len(data) # sample size
mean = sum(data) / n # sample mean
stddev = 12 # population standard deviation

t_value = stats.t.ppf(0.95, n-1)
margin_error = t_value * (stddev / (n ** 0.5))
confidence_interval = (mean - margin_error, mean + margin_error)

print("90% Confidence Interval:", confidence_interval)
```

# Q6

```python
"""
Use Python to plot the chi-square distribution with 10 degrees of freedom. 
Label the axes and shade the area corresponding to a chi-square statistic of 15.
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

df = 10 # degrees of freedom
x = np.linspace(0, 30, 1000)
y = chi2.pdf(x, df)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(x, y, label='Chi-Square PDF')
ax.fill_between(x[x >= 15], y[x >= 15], 0, alpha=0.5, label='Area: Chi-Square > 15')
ax.set_title(f'Chi-Square Distribution with {df} Degrees of Freedom')
ax.set_xlabel('Chi-Square Statistic')
ax.set_ylabel('Probability Density')
ax.legend()
plt.show()
```
