```python
"""
Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose  you have collected data on the amount of time students spend studying for an exam and their final exam  scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.
"""
```

```python
"""
Let's assume you have collected data for 30 students. Here's an example set of data:

Student 1: Time studying (hours) = 10, Exam score = 85
Student 2: Time studying (hours) = 8, Exam score = 78
...
Student 30: Time studying (hours) = 12, Exam score = 92

The formula for Pearson correlation coefficient (r) is:
r = (Σ((x - x̄)(y - ȳ))) / sqrt(Σ(x - x̄)² * Σ(y - ȳ)²)

For example, if the calculated Pearson correlation coefficient is 0.75, it would indicate a strong positive linear relationship between the amount of time spent studying and the final exam scores. This means that as the time spent studying increases, the exam scores tend to increase as well.
"""
```

```python
"""
Spearman's rank correlation is a measure of the monotonic relationship between two variables.  Suppose you have collected data on the amount of sleep individuals get each night and their overall job  satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two  variables and interpret the result.
"""
```

```python
"""
Let's assume you have collected data for 50 individuals. Here's an example set of data:

Individual 1: Sleep (hours) = 7, Job satisfaction rank = 8
Individual 2: Sleep (hours) = 6, Job satisfaction rank = 6
...
Individual 50: Sleep (hours) = 8, Job satisfaction rank = 9

The Spearman's rank correlation coefficient using the formula:
ρ = 1 - ((6 * Σd²) / (n³ - n))

For example, if the calculated Spearman's rank correlation coefficient is 0.85, it would indicate a strong positive monotonic relationship between the amount of sleep and job satisfaction. This means that as the amount of sleep increases, job satisfaction tends to increase as well.
"""
```

```python
"""
The Pearson correlation coefficient:
r = Σ((x - x̄)(y - ȳ)) / sqrt(Σ(x - x̄)² * Σ(y - ȳ)²)

The Spearman's rank correlation coefficient:
ρ = 1 - ((6 * Σd²) / (n³ - n))

Let's assume you have collected the following data:

Participant 1: Hours of exercise per week = 4, BMI = 25.2
Participant 2: Hours of exercise per week = 3, BMI = 27.5
...
Participant 50: Hours of exercise per week = 5, BMI = 24.8

Step 1: Calculate the means:
Mean of hours of exercise per week (x̄) = (4 + 3 + ... + 5) / 50
Mean of BMI (ȳ) = (25.2 + 27.5 + ... + 24.8) / 50

Step 2: Calculate the Pearson correlation coefficient (r):
Calculate the sum of ((x - x̄)(y - ȳ)), the sum of (x - x̄)², and the sum of (y - ȳ)² using the given data. Then substitute the values into the formula for r.

Step 3: Calculate the Spearman's rank correlation coefficient (ρ):
Rank the data for each variable (hours of exercise and BMI). Calculate the sum of squared differences between the ranks (Σd²). Then substitute the values into the formula for ρ.

The correlation coefficients will be between -1 and 1. A positive value indicates a positive relationship, while a negative value indicates a negative relationship. A value close to 0 suggests a weak or no relationship.
"""
```

```python
"""
A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.
"""
```

```python
import numpy as np

# Sample data for sales calls per day and sales per week
sales_calls = [20, 15, 25, 18, 22, 17, 19, 16, 21, 23, 18, 24, 16, 20, 19, 15, 21, 23, 17, 22, 19, 25, 18, 16, 20, 15, 21, 23, 17, 22]
sales_per_week = [50, 45, 60, 52, 58, 46, 50, 48, 55, 60, 52, 59, 47, 51, 49, 45, 54, 59, 47, 56, 49, 60, 51, 47, 51, 44, 55, 57, 48, 58]

# Calculate the Pearson correlation coefficient
correlation_coefficient = np.corrcoef(sales_calls, sales_per_week)[0, 1]

print("Pearson correlation coefficient:", correlation_coefficient)
```
