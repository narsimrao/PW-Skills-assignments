```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, output_file, show 
from bokeh.sampledata.iris import flowers
```

```python
flowers
```

# Q1

```python
"""
How can you create a Bokeh plot using Python code?
"""
```

```python
output_file("test.html")

p = figure(title = "test flower")
p.xaxis.axis_label = "petal_length"
p.yaxis.axis_label = "petal_width"

p.circle(flowers["petal_length"],flowers["petal_width"])
show(p)
```

# Q2

```python
"""
What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
"""
```

```python
"""
a glyph is a visual shape used to represent data on a plot. 
It can be any shape or marker such as a circle, square, line, or triangle.

Glyphs are added to a Bokeh plot using the figure() function
"""
output_file("test.html")
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

p = figure()
p.circle(x, y,size=10, color='red')
show(p)
```

# Q3

```python
"""
How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?
"""
```

```python
output_file("test.html")

p = figure(plot_width=800, plot_height=400)

x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

p.xaxis.axis_label = 'X-axis Label'
p.xaxis.axis_label_text_font_size = '14pt'

p.title.text = 'Title'
p.title.text_font_size = '16pt'

p.legend.label_text_font_size = '12pt'
p.legend.location = 'top_left'

p.circle(x, y,size=10, color='red')
show(p)
```
