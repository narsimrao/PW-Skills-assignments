```python
"""
url_for() function is used for URL building in flask.
"""

from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return '<h1>Company Name: ABC Corporation<br/>Location: India<br/>Contact Detail: 999-999-9999</h1>'

@app.route('/welcome')
def welcome():
    return '<h1>Welcome to ABC Corporation</h1>'

@app.route('/cricket')
def cricket():
    return '<h1>This is cricket page</h1>'

@app.route('/football')
def football():
    return '<h1>This is football page</h1>'

@app.route('/hockey')
def hockey():
    return '<h1>This is hockey page</h1>'

@app.route('/game/<name>')
def game(name):
    if name == 'cricket':
        return redirect(url_for('cricket'))
    if name == 'football':
        return redirect(url_for('football'))
    if name == 'hockey':
        return redirect(url_for('hockey'))


if __name__=='__main__':
    app.run(host='0.0.0.0')
```
