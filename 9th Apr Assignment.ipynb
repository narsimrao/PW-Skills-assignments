# Q2

```python
"""
What is the formula for Bayes' theorem?
"""
```

```python
"""
P(A|B) = (P(B|A) * P(A)) / P(B)

where:

P(A|B) represents the probability of event A given event B has occurred.
P(B|A) represents the probability of event B given event A has occurred.
P(A) and P(B) represent the probabilities of events A and B, respectively.
"""
```

```python
"""
Bayes' theorem is based on conditional probability. It provides a way to calculate the conditional probability of an event A given event B, by using the conditional probability of event B given event A along with the individual probabilities of events A and B.
"""
```
